```python
import pandas as pd

years = list(range(2015, 2026))
base_url = "https://raw.githubusercontent.com/Tennismylife/TML-Database/master/{}.csv"

dfs = []
for year in years:
    url = base_url.format(year)
    try:
        df = pd.read_csv(url)
        df["year"] = year
        dfs.append(df)
        print(f"{year} cargado")
    except Exception as e:
        print(f"Error en {year}: {e}")

all_matches = pd.concat(dfs, ignore_index=True)
```

```python
# Ver las dimensiones del dataset
print(f"Número total de partidos: {all_matches.shape[0]}")
print(f"Número de columnas: {all_matches.shape[1]}")

# Mostrar las primeras filas
display(all_matches.head())

# Ver nombres de las columnas
print("Columnas disponibles:")
print(all_matches.columns.tolist())

# Ver resumen general
display(all_matches.describe(include='all'))

# Verificar valores nulos por columna
print("\nValores nulos por columna:")
print(all_matches.isnull().sum())
```

```python
# Asegurarse de que las columnas son numéricas y manejar posibles valores no numéricos/nulos
all_matches['winner_rank'] = pd.to_numeric(all_matches['winner_rank'], errors='coerce')
all_matches['loser_rank'] = pd.to_numeric(all_matches['loser_rank'], errors='coerce')

# Eliminar filas donde cualquiera de las columnas sea NaN después de la conversión
matches_with_ranks = all_matches.dropna(subset=['winner_rank', 'loser_rank'])

# Contar el número de filas donde winner_rank es menor que loser_rank
winner_lower_rank_count = (matches_with_ranks['winner_rank'] < matches_with_ranks['loser_rank']).sum()

# Calcular el porcentaje
total_matches_with_ranks = matches_with_ranks.shape[0]
percentage_winner_lower_rank = (winner_lower_rank_count / total_matches_with_ranks) * 100

print(f"Porcentaje de filas donde el winner_rank es menor que el loser_rank: {percentage_winner_lower_rank:.2f}%")
```
