```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/alzatem204/Tennis-Predictions/main/General_Data_ATP_UK.csv"
response = requests.get(url)

with open("General_Data_ATP_UK.csv", "wb") as f:
    f.write(response.content)

all_matches = pd.read_csv("General_Data_ATP_UK.csv")
display(all_matches.head())
```

```python
# Ver las dimensiones del dataset
print(f"Número total de partidos: {all_matches.shape[0]}")
print(f"Número de columnas: {all_matches.shape[1]}")

# Mostrar las primeras filas
display(all_matches.head())

# Ver nombres de las columnas
print("Columnas disponibles:")
print(all_matches.columns.tolist())

# Ver resumen general
display(all_matches.describe(include='all'))

# Verificar valores nulos por columna
print("\nValores nulos por columna:")
print(all_matches.isnull().sum())
```

```python
# Eliminar filas con valores nulos en WRank o LRank para asegurar la comparación
df_cleaned = all_matches.dropna(subset=['WRank', 'LRank']).copy()

# Convertir WRank y LRank a tipo numérico, coercing errors to NaN
df_cleaned['WRank'] = pd.to_numeric(df_cleaned['WRank'], errors='coerce')
df_cleaned['LRank'] = pd.to_numeric(df_cleaned['LRank'], errors='coerce')

# Eliminar filas donde la conversión a numérico resultó en NaN
df_cleaned = df_cleaned.dropna(subset=['WRank', 'LRank'])

# Calcular el número de filas donde WRank es mayor que LRank
count_w_rank_greater = df_cleaned[df_cleaned['WRank'] < df_cleaned['LRank']].shape[0]

# Calcular el porcentaje
percentage = (count_w_rank_greater / df_cleaned.shape[0]) * 100 if df_cleaned.shape[0] > 0 else 0

print(f"El porcentaje de filas donde WRank es mayor que LRank es: {percentage:.2f}%")
```
